```python
# Additional dependencies for modeling
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import MiniBatchKMeans
from sklearn.impute import SimpleImputer
```

```python
# Baseline pizza slice from gold search table
pizza_mask = gold_food_search['description'].str.contains('pizza', case=False, na=False)
baseline_results = gold_food_search[pizza_mask].copy()
print(f"Baseline pizza hits: {len(baseline_results):,}")
baseline_results[['fdc_id', 'description', 'brand_owner', 'data_type', 'branded_food_category']].head()
```

```python
# Train a simple logistic model to score "general" foods (non-branded) for ranking
training_df = gold_food_search.copy()
training_df['is_general'] = training_df['data_type'].isin(['foundation_food', 'sr_legacy_food', 'survey_fndds_food'])

feature_cols_numeric = ['serving_size', 'kcal', 'protein_g', 'fat_g', 'carb_g']
feature_cols_categorical = ['data_type', 'branded_food_category']

numeric_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value=0.0)),
                                      ('scaler', StandardScaler(with_mean=False))])
categorical_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
                                          ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, feature_cols_numeric),
        ('cat', categorical_transformer, feature_cols_categorical),
    ]
)

ranker = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('model', LogisticRegression(max_iter=1000))
])

ranker.fit(training_df[feature_cols_numeric + feature_cols_categorical], training_df['is_general'])
baseline_results['general_score'] = ranker.predict_proba(baseline_results[feature_cols_numeric + feature_cols_categorical])[:, 1]
pizza_ranked = baseline_results.sort_values('general_score', ascending=False)
pizza_ranked[['fdc_id', 'description', 'data_type', 'general_score']].head(10)
```

```python
# TF-IDF semantic embeddings for descriptions + ingredients
text_corpus = (gold_food_search['description'].fillna('') + ' ' + gold_food_search['ingredients'].fillna(''))
vectorizer = TfidfVectorizer(stop_words='english', min_df=5)
text_vectors = vectorizer.fit_transform(text_corpus)

def semantic_search(query: str, top_k: int = 10):
    query_vec = vectorizer.transform([query])
    sims = cosine_similarity(query_vec, text_vectors).ravel()
    top_idx = sims.argsort()[::-1][:top_k]
    return gold_food_search.iloc[top_idx][['fdc_id', 'description', 'brand_owner', 'data_type', 'branded_food_category']].assign(similarity=sims[top_idx])

semantic_results = semantic_search('pizza', top_k=10)
semantic_results
```

```python
# Cluster pizza items to identify near-duplicate groups
pizza_text = (baseline_results['description'].fillna('') + ' ' + baseline_results['brand_owner'].fillna(''))
pizza_vectors = vectorizer.transform(pizza_text)

n_clusters = 20
cluster_model = MiniBatchKMeans(n_clusters=n_clusters, random_state=42, batch_size=2048)
cluster_labels = cluster_model.fit_predict(pizza_vectors)
baseline_results['cluster_id'] = cluster_labels
cluster_sizes = baseline_results.groupby('cluster_id').size().sort_values(ascending=False)
cluster_sizes.head(10)
```

```python
# Inspect the largest cluster to see representative items
largest_cluster = cluster_sizes.index[0]
cluster_preview = baseline_results[baseline_results['cluster_id'] == largest_cluster]
cluster_preview[['fdc_id', 'description', 'brand_owner', 'cluster_id']].head(10)
```

```python
# Intent/category suggestion model using multiclass logistic regression
category_counts = gold_food_search['branded_food_category'].fillna('unknown').value_counts()
top_categories = category_counts[category_counts > 5000].index.tolist()

intent_df = gold_food_search.copy()
intent_df['category_target'] = intent_df['branded_food_category'].fillna('unknown')
intent_df.loc[~intent_df['category_target'].isin(top_categories), 'category_target'] = 'other'

intent_vectorizer = TfidfVectorizer(stop_words='english', min_df=10)
intent_X = intent_vectorizer.fit_transform(intent_df['description'].fillna(''))
intent_model = LogisticRegression(max_iter=1000, multi_class='ovr')
intent_model.fit(intent_X, intent_df['category_target'])

def suggest_categories(query: str, top_k: int = 5):
    vec = intent_vectorizer.transform([query])
    probs = intent_model.predict_proba(vec)[0]
    top_idx = np.argsort(probs)[::-1][:top_k]
    return [(intent_model.classes_[i], probs[i]) for i in top_idx]

suggest_categories('pizza')
```

```python
# Personalisation example: rank pizzas against macro targets
macro_targets = {
    'high_protein': {'protein_g': 25, 'kcal': 600},
    'low_carb': {'carb_g': 20, 'kcal': 500},
}

def macro_score(row, target):
    score = 0.0
    if 'protein_g' in target and row['protein_g'] > 0:
        score += min(row['protein_g'] / target['protein_g'], 2.0)
    if 'carb_g' in target and row['carb_g'] > 0:
        score += max(0, 1 - row['carb_g'] / target['carb_g'])
    if 'kcal' in target and row['kcal'] > 0:
        score += max(0, 1 - abs(row['kcal'] - target['kcal']) / target['kcal'])
    return score

personalised_views = {}
for goal, target in macro_targets.items():
    scores = baseline_results.apply(lambda row: macro_score(row, target), axis=1)
    personalised_views[goal] = baseline_results.assign(personal_score=scores).sort_values('personal_score', ascending=False).head(5)[['fdc_id', 'description', 'protein_g', 'carb_g', 'kcal', 'personal_score']]

personalised_views
```
